```python
import numpy as np
import matplotlib.pyplot as plt

# Create Directory and Add Science Styles
plt.style.use(['science', 'ieee'])
plt.rcParams["text.usetex"] = False
plt.rcParams['figure.figsize'] = 6, 2
```

```python
values = [0.925826221, 0.978632595, 0.999632595, 0.965826221]
plt.plot(values, '--.', c='0.35')
plt.ylabel('F1 Score')
plt.xlabel('Dataset Ratio')
plt.xticks(range(len(values)), ['20%', '40%', '60%', '80%'])
plt.legend()
plt.show()
```

```python
values = [0.935826221, 0.998632595, 0.989632595, 0.965826221]
plt.plot(values, '--g.')
plt.ylabel('F1 Score')
plt.xlabel('Window Size')
plt.xticks(range(len(values)), ['5', '10', '15', '20'])
plt.legend()
plt.show()
```

```python
centralized_loss = [0.5, 0.4, 0.2, 0.15, 0.1, 0.08, 0.06, 0.05, 0.04, 0.03, 0.02, 0.015]
federated_loss = [0.8, 0.3, 0.1, 0.05, 0.02, 0.01, 0.005, 0.002, 0.001, 0.0005, 0.0002, 0.0001]
plt.plot(centralized_loss, '--b.', label='Centralized Loss')
plt.plot(federated_loss, '--r.', label='Federated Loss')
plt.ylabel('Average Trianing Loss')
plt.xlabel('Epochs')
plt.legend()
plt.show()
```
